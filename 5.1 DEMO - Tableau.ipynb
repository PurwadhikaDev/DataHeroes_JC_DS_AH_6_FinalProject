# Demo RandomForest - `Penentuan Harga Pasar`

## Import Data Regression

```python
data_reg = pd.read_csv('Data/DC_Prop_Residential_Regression_DEMO1.csv', index_col= 'Unnamed: 0')

# Set Index
indeks = pd.Series(range(data_reg.shape[0]))
data_reg = data_reg.set_index([indeks])

data_reg_RF = data_reg.drop(['ID_PROP', 'PRICE'], axis =1)
```

```python
data_reg_RF.head()
```

```python
model_RF = 'model_randomForest_r_undrop.sav'
loaded_model_RF = pickle.load(open(model_RF, 'rb'))
```

```python
predict_RF= pd.DataFrame(loaded_model_RF.predict(data_reg_RF))
predict_RF.columns = ['PREDICT_PRICE']
predict_RF['PREDICT_PRICE'] = predict_RF['PREDICT_PRICE']**3
predict_RF
```

```python
cols = data_reg.columns.to_list() + predict_RF.columns.to_list()
dfs = [data_reg,predict_RF]
result = np.concatenate(dfs, axis=1)  
result = pd.DataFrame(result, columns=cols)
result.head()
```

# Demo KMEANS - `Penentuan Resiko`

## Import data Clustering

```python
# Loading the dataset
data_clus = pd.read_csv('Data/DC_Prop_Residential_Clustering_DEMO1.csv', index_col='Unnamed: 0')
```

```python
data_clus.shape
```

```python
# Set Index
indeks = pd.Series(range(data_clus.shape[0]))
data_clus = data_clus.set_index([indeks])

# Sampling
data_clus_KM = data_clus.drop(['ID_PROP'], axis =1)
```

```python
model_KM = 'model_Clus_Kmeans_r.sav'
loaded_model_KM = pickle.load(open(model_KM, 'rb'))
```

```python
predict_KM= pd.DataFrame(loaded_model_KM.predict(data_clus_KM))
predict_KM.columns = ['PREDICT_CLUSTER']
predict_KM.shape
```

```python
cols = data_clus.columns.to_list() + predict_KM.columns.to_list()
dfs = [data_clus,predict_KM]
result_data_clus = np.concatenate(dfs, axis=1)  
result_data_clus = pd.DataFrame(result_data_clus, columns=cols)
result_data_clus.shape
```

```python
result_data_clus['RISK'] = np.where(result_data_clus['PREDICT_CLUSTER'] == 0, 0.8, 
                    (np.where(result_data_clus['PREDICT_CLUSTER'] == 1, 0.5, 0.7)))
```

```python
result_data_clus
```

```python
cluster_result_only = result_data_clus[['ID_PROP', 'PREDICT_CLUSTER',	'RISK']]
```

```python
result = result.set_index('ID_PROP').join(cluster_result_only.set_index('ID_PROP'))
```

## PREDIKSI HARGA BELI

```python
result['HARGA_BELI']=result['PREDICT_PRICE']*result['RISK']
result
```

## DEMO TIME SERIES - `Prediksi Tren Kenaikan Harga`

```python
tren_res = 0.0681
tren_con = 0.1643

result['TREND'] = tren_res
```

## Prediksi Harga Jual

```python
result['HARGA_JUAL'] = result['PREDICT_PRICE'] * (1+result['TREND'])
result['PROFIT'] = result['HARGA_JUAL'] - result['PREDICT_PRICE']

AYH = result[[ 'PREDICT_PRICE',	'PREDICT_CLUSTER',	'RISK',	'HARGA_BELI', 'TREND', 'HARGA_JUAL', 'PROFIT']]
AYH
```
